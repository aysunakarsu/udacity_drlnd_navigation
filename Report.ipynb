# Project 1: Navigation

###  Author Aysun Akarsu

##  Objective

This project's aim is to train a reinforcement learning agent to navigate (and collect bananas!) in a restricted environment. A reward of +1 is provided for collecting a yellow banana, and a reward of -1 is provided for collecting a blue banana.  Thus, the goal of the agent is to collect as many yellow bananas as possible while avoiding blue bananas.  

![Trained Agent][image1]

The task is episodic, and in order to solve the environment, the agent must get an average score of +13 over 100 consecutive episodes.

## Learning Algorithm

In this project deep Q-learning algorithm is used to train the agent by implementing a *Value Based* method called [Deep Q-Networks](https://deepmind.com/research/dqn/).

Deep Q Learning combines 2 approaches :
- A Reinforcement Learning method called [Q Learning](https://en.wikipedia.org/wiki/Q-learning) (aka SARSA max)
- A Deep Neural Network to learn a Q-table approximation (action-values)
  
### Hyper Parameters  
The agent uses the following hyper parameters<br>
BUFFER_SIZE = int(1e5)  # replay buffer size<br>
BATCH_SIZE = 64         # minibatch size<br>
GAMMA = 0.99            # discount factor<br>
TAU = 1e-3              # for soft update of target parameters<br>
LR = 5e-4               # learning rate<br>
UPDATE_EVERY = 4        # how often to update the network<br>


### The QNetwork
  Fully connected layer1: Linear(in_features=37, out_features=64, bias=True)<br>
  Fully connected layer2: Linear(in_features=64, out_features=64, bias=True)<br>
  Fully connected layer3: Linear(in_features=64, out_features=64, bias=True)<br>
  Fully connected layer4: Linear(in_features=64, out_features=4, bias=True)<br>
